```python
dados.plot(x="Unidade da Federação", y = "2008/Ago", kind="bar",figsize=(9,6))
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = dados.plot(x="Unidade da Federação", y = "2008/Ago", kind="bar",figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
ticker.FormatStrFormatter
plt.title("Valor por unidade da Federação")
plt.show()
```

##Desafio: faça a mesma análise para um mês mais recente

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = dados.plot(x="Unidade da Federação", y = "2021/Mar", kind="bar",figsize=(10,7),xlabel = "Estados", ylabel = "Valor (R$)", fontsize=7.5)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
ticker.FormatStrFormatter
plt.title("VALOR TOTAL DE PROCEDIMENTOS HOSPITALARES DO SUS")
plt.show()
```
